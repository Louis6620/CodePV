```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import stats
from datetime import datetime
dateparse = lambda x: datetime.strptime(x, '%m/%d/%Y %H:%M')
import plotly.express as px
import plotly.graph_objects as go
from mpl_toolkits import mplot3d
```

```python
df = pd.read_csv('./t345678.csv',
                encoding="utf-8-sig",
                header=0,
                infer_datetime_format=True,
                parse_dates={'datetime':[0]},
                index_col=['datetime']
                )
```

```python
df.head()
#xem lại file csv
```

```python
df=df[['Ambient temperature(℃)', 'PV Temperature(℃)', 'Daily irradiation(MJ/㎡)', 'Irradiance(W/㎡)', 'Inverter efficiency(%)', 'Active power(kW)', 'Daily energy(kWh)', 'Total input power(kW)', 'Cumulative energy(kWh)','Inverter startup time','Inverter shutdown time']].dropna()
df.head()
#Chọn những trường sử dụng 
#Xem lại dữ liệu của dataframe
```

```python
#Trích xuất thời gian đơn giản hơn và xác định ngày tháng tuần theo thời gian cho trước
df['time'] = df.index
df.loc[:,'month'] = pd.Series(df.index.month, df.index)
df.loc[:,'day'] = pd.Series(df.index.day, df.index)
df.loc[:,'hour'] = pd.Series(df.index.hour, df.index)
df.loc[:,'dayofweek'] = pd.Series(df.index.dayofweek, df.index)
df.loc[:,'dayname'] = pd.Series(df.index.day_name(), df.index)
df.loc[:,'hourofweek'] = pd.Series((df.hour+24*df.dayofweek), df.index)
df.loc[:,'hourofmonth'] = pd.Series((df.hour+24*(df.day-1)), df.index)
##
df.loc[:,'Tcell']=pd.Series((df['Ambient temperature(℃)']+df['Irradiance(W/㎡)']/800*(25-20)), df.index) #Tính toán nhiệt độ tấm pin Tcell bằng nhiệt độ môi trường và bức xạ từ dữ liệu của weaterstation 
df.loc[:,'Pm']=pd.Series((2.172*1.303*220*2*590*df['Irradiance(W/㎡)']/1000*(1-0.0045*(df['Tcell']-25))*df['Inverter efficiency(%)']/100*(98-0.55*5/12)/100*20.1/100), df.index) #Tính toán hiệu suất
df.loc[:,'Hieu suat']=pd.Series((df['Total input power(kW)']*1000/df['Pm']*100), df.index) #Tính toán tỷ suất 
```

```python
#xem lại sự thay đổi của dataframe
df.head()
```

```python
#thay thế những NaN chưa được xử lý ở trên bằng 0 và kiểm tra lại sự thay đổi
df=df.replace(np.nan,0)
df.head()
```

```python
#trích xuất thành file csv mới
df.to_csv('test.csv')
```

```python
gio=[]
tgian=[]
dem=0
t=0
tg=0
tong=0
for i in range(len(df['time'])-1):
    dem=dem+1
    tong=tong+df['Irradiance(W/㎡)'][i]
    if df['hour'][i]!=df['hour'][i+1]:
        gio.append(tong/dem)
        if df['hourofmonth'][i]>tg:
            t=df['hourofmonth'][i]-tg+t
        else:
            t=df['hour'][i]+24-df['hour'][i-dem]+t
        tgian.append(t)
        tg=df['hourofmonth'][i]
        dem=0
        tong=0
tong=tong+df['Irradiance(W/㎡)'][i+1]
if df['hourofmonth'][i]>tg:
    t=df['hourofmonth'][i]-tg+t
else:
    t=df['hour'][i]+24-df['hour'][i-dem]+t
tgian.append(t)
gio.append(tong/(dem+1))
```

```python
print(tgian)
print(gio)
```

```python
fig = go.Figure()
fig.add_trace(go.Bar(
    x=tgian,
    y=gio,
    name="gio trung binh",
    yaxis="y",
))
fig . update_traces (
    showlegend = True)
fig.update_xaxes(
    rangeslider_visible=True,
)
fig.show()
```
